```python
initial = [(0, 1), (0, 4), (0, 5), (1, 3), (2, 1), (3, 2), (3, 4)]
```

## Adjacency list should contain all nodes as keys even if they have to outcoming vertexes

```python
def construct_adjacency_list(initial):
    graph = {}
    for pair in initial:
        try:
            graph[pair[0]].append(pair[1])
        except KeyError:
            graph[pair[0]] = [pair[1]]
        try:
            graph[pair[1]]
        except KeyError:
            graph[pair[1]] = []
    return graph
```

```python
test_graph = construct_adjacency_list(initial)
```

```python
print (test_graph)
```

## Depth-first search

```python
def depth_first_search(adj_list, start, visited=[]):
    visited.append(start)
    for node in adj_list[start]:
        if node in adj_list:
            if node not in visited:
                depth_first_search(adj_list, node, visited)
    return visited
```

```python
depth_first_search(test_graph, 0)
```

# Breadth-first search

```python
Image('./directed_graph_bfs_dfs.png')
```

```python
import queue
```

```python
def breadth_first_search(adj_list, start, visited=[]):
    q = queue.Queue()
    visited.append(start)
    q.put(start)
    while not q.empty():
        cur_node = q.get()
        for node in adj_list[cur_node]:
            if node not in visited:
                q.put(node)
                visited.append(node)
    return visited
```

```python
breadth_first_search(test_graph, 0)
```
